```python
# 
# input (num1, num2)
# output (factor, multiple, neither)

input_line = []
input_list = list(map(int,input().split()))

for i in range(int(len(input_list)/2)-1):
  num1 = input_list[2*i]
  num2 = input_list[1+2*i]

  if (num1 == 0)&(num2 ==0):
    break
  else:
    if num1 > num2 :
      div = num1%num2
      if div == 0:
        print("multiple")
      else:
        print("neither")

    if num2 > num1:
      div = num2%num1
      if div == 0:
        print("factor")
      else:
        print("neither") 
```
