```python
## DataStudio에 맞게 데이터 변환하기

# 데이터 불러오기
import os, sys
from google.colab import drive
import pandas as pd
```

```python
drive.mount('/content/drive')
```

```python
origin_dt = pd.read_csv('/content/drive/MyDrive/Data/datastudio/gdp_data.csv')
```

```python
origin_dt
```

```python
year_list = origin_dt.columns[3:]

year_list
```

```python
data= origin_dt.melt(id_vars=['Country Name','Country Code'], value_vars=year_list)
```

```python
# Nan 값 제거
data = data.dropna()
```

```python
# 칼럼 명 변경
data.columns = ['Country','Contry Code','Year','GDP(billion$)']

data
```

```python
# 변경된 데이터 저장
data.to_csv("/content/drive/MyDrive/Data/datastudio/t_gdp_data.csv",header=True, index=None)
```
